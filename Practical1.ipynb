```python
plt.figure
plt.hist(df['mpg'], bins=10)
plt.xlabel("Miles Per Gallon (mpg)")
plt.ylabel("Frequency")
plt.title("Rohan's First Practical")
plt.show()
```

```python
#compute histo values
counts, bin_edges = np.histogram(df['mpg'], bins=10)

#find idx of highest freq
max_index = np.argmax(counts)

#get interval
highest_interval = (bin_edges[max_index], bin_edges[max_index + 1])

print("highest freq:", counts[max_index])
print("highest freq interval:", highest_interval)
```
